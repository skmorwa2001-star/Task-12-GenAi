# Assignment 12: Seaborn (Realational , Distribution, Categorical & Multi-Plots)

```python
# Import labraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Task 1: Relational Plot

```python
# Read csv file
data=pd.read_csv('Insurance.csv')
```

```python
# Create dataframe as df
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
## Create a relational plot (relplot) 

sns.relplot(data=df,x='Age',y='Policy Term',kind='line',hue='Plan')
plt.show()
```

```python
## Create the same plot using scatter style

sns.relplot(data=df,x='Age',y='Policy Term',hue='Plan')
plt.show()
```

## Task 2: Line Plot as Scatter & Facet

```python
df.head()
```

```python
# Create a line plot using ----> sns.lineplot()
plt.title('Relationship between Amount and Age')
sns.lineplot(data=df,x='Age',y='Amount')
plt.show()
```

```python
# Convert the same relationship into a scatter style line plot
plt.title('Relationship between Amount and Age')
sns.scatterplot(data=df,x='Age',y='Amount',hue='Plan')
plt.show()
```

```python
# Use faceting (col or row ) to split the plot based on a categorical column
sns.relplot(data=df,x='Age',y='Amount',col_wrap=2,kind='scatter',col='Plan')
plt.show()
```

## Task 3: Distribution Plots

```python
df.head()
```

```python
# Plot Histogram usin Seaborn

sns.histplot(data=df,x='Age',bins=20)
plt.show()
```

```python
## Plot KDE plot

sns.kdeplot(data=df,x='Age',hue='Plan',fill=True)
plt.grid()
plt.show()
```

```python
### Plot rug plot

sns.rugplot(data=df,x='Age')
plt.show()
```

```python
#### Combine histogram + KDE in a single plot

sns.histplot(data=df,x='Age',bins=20,kde=True)
plt.grid()
plt.show()
```

## Task 4: Bivariate Distribution Plots

```python
df.head()
```

```python
# Create a bivariate histogram

sns.histplot(data=df,x='Age',y='Amount')
plt.show()
```

```python
## Create a bivariate KDE plot

sns.kdeplot(data=df,x='Age',y='Amount')
plt.show()
```

## Task 5: Matrix Plots

```python
df.head()
```

```python
# Create a pair plot 

sns.pairplot(data=df)
plt.show()
```

```python
# Use for correlation
corr = df.select_dtypes(include='number').corr()
```

```python
## Create a heatmap of correlation matrix

sns.heatmap(data=corr,annot=True,cmap='coolwarm')
plt.show()
```

## Task 6: Categorical Plots

```python
df.head()
```

```python
# Bar plot

sns.barplot(data=df,x='Plan',y='Age')
plt.show()
```

```python
## Box plot

sns.boxplot(data=df,x='Amount')
plt.show()
```

```python
### Violin plot

sns.violinplot(data=df,x='Plan',y='Age')
plt.show()
```

```python
#### Count Plot

sns.countplot(data=df,x='Policy Term')
plt.show()
```

## Task 7: Regression Plots

```python
# Create a regression plot (regplot) between two numerical columns

sns.regplot(data=df,x='Age',y='Amount')
plt.show()
```

```python
## Create an implot with hue using a categorical column

sns.lmplot(data=df,x='Age',y='Amount',hue='Plan')
plt.show()
```

## Task 8: Multi-Plots & Figure-Level Plots

```python
df.head()
```

```python
# Create a facetgrid

g=sns.FacetGrid(data=df,col='Plan',col_wrap=2)
g.map(sns.scatterplot,'Age','Amount')
plt.show()
```

```python
sns.displot(data=df,x='Age',col='Plan',col_wrap=2,kde=True)
plt.show()
```
